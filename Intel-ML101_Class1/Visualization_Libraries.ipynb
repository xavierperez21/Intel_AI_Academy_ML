<h3>Visualizations can be created in multiple ways:</h3>

- Matplotlib
- Pandas (via Matplotlib)
- Seaborn
    - Statistically-focused plotting methods
    - Global preferences incorporated by Matplotlib

<h2 style="color: green">BASIC SCATTER PLOTS WITH MATPLOTLIB</h3>

<h3>Scatter plots can be created from Pandas Series</h3>

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#The location of the data file
filepath = 'data/Iris_Data.csv'

#Import the data
data = pd.read_csv(filepath)
```

```python
plt.plot(data.sepal_length,
         data.sepal_width,
         ls = '',
         marker='o')
plt.show()
```

<h3>Multiple layers of data can also be added</h3>

```python
plt.plot(data.sepal_length, 
         data.sepal_width,
         ls = '', marker='o',
         label = 'sepal')
plt.plot(data.petal_length,
         data.petal_width,
         ls = '', marker='o',
         label='petal')
plt.show()
```

<h3>Histograms can be created from Pandas Series</h3>

```python
plt.hist(data.sepal_length, bins=25)
plt.show()
```

<h2 style="color: orange">CUSTOMIZING MATPLOTLIB PLOTS</h3>

<h3>Every feature of Matplotlib plots can be customized</h3>

```python
fig, ax = plt.subplots()

ax.barh(np.arange(10), data.sepal_width.iloc[:10])

#Set position of ticks and tick labels
ax.set_yticks(np.arange(0.4,10.4,1.0))
ax.set_yticklabels(np.arange(1,11))
ax.set(xlabel='xlabel', ylabel='ylabel', title='Title')

plt.show(ax)
```

<h2 style="color: red">INCORPORATING STATISTICAL CALCULATIONS</h3>

<h3>Statistical calculations can be included with Pandas methods</h3>

```python
plt.show(data.groupby('species').mean().plot(color=['red',
                                             'blue',
                                             'black',
                                             'green'],
                                      fontsize=10.0,
                                      figsize=(4,4)))
```

<h2 style="color: #A0F">STATISTICAL PLOTTING WITH SEABORN</h3>

<h3>Joint distribution and scatter plots can be created</h3>

```python
import seaborn as sns
```

```python

sns.jointplot(x='sepal_length',
              y='sepal_width',
              data=data, height=4)
```

```python
sns.pairplot(data, hue='species', height=3)
```
